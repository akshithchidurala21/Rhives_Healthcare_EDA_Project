## Exploratory Data Analysis (EDA) & Visualization of Trends

```python
cols = {c.lower(): c for c in df.columns}

def find_col(key_words):
    """Return the first column name that contains ALL keywords (case-insensitive)."""
    for lower_name, orig in cols.items():
        if all(k in lower_name for k in key_words):
            return orig
    return None

year_col = find_col(['year']) or find_col(['yr'])
deaths_col = find_col(['death']) or find_col(['deaths']) or find_col(['number of deaths'])
country_col = find_col(['country']) or find_col(['territory'])
region_col = find_col(['region'])
rate_col = find_col(['rate']) or find_col(['case detection rate']) or find_col(['per 100 000']) or find_col(['per 100000'])
pop_col = find_col(['population']) or find_col(['pop'])

print("Detected columns:")
print(" Year:", year_col)
print(" Deaths:", deaths_col)
print(" Country:", country_col)
print(" Region:", region_col)
print(" Rate:", rate_col)
print(" Population:", pop_col)
```

### a. Distribution of Deaths Over the Years

```python
deaths_over_year = df.groupby("Year")["Estimated number of deaths from TB (all forms, excluding HIV)"].sum()
deaths_over_year
```

```python
plt.figure(figsize=(8,6))
plt.plot(deaths_over_year.index, deaths_over_year.values, marker='o', color='skyblue')
plt.title("Distribution of TB Deaths Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Deaths")
plt.grid(True)
plt.show()
```

### b. Top 10 Countries with the Highest Death Rates

```python
top10= df.groupby("Country or territory name")["Estimated number of deaths from TB (all forms, excluding HIV)"].sum().sort_values(ascending=False).head(10)
top10
```

```python
plt.figure(figsize=(8,6))
top10.plot(kind='bar', x='Country', y='Deaths', legend=False, color='green')
plt.title("Top 10 Countries with the Highest TB Deaths")
plt.xlabel("Country")
plt.ylabel("Number of Deaths")
plt.xticks
plt.tight_layout()
plt.savefig("A:\Rhives\Week-3/top10_tb_deaths.png", dpi=150)
plt.show()
```

### c)Deaths per Population (per 100k)

```python
deaths_col = "Estimated number of deaths from TB (all forms, excluding HIV)"
pop_col = "Estimated total population number"
region_col = "Region"

df["Deaths_per_100k"] = (df[deaths_col] / df[pop_col]) * 100000
region_stats = df.groupby(region_col)["Deaths_per_100k"].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(5,5))
plt.pie(
    region_stats,
    labels=region_stats.index,
    autopct='%1.1f%%',
    startangle=140,
    colors = ['#A2D2FF', '#BDE0FE', '#C1E1C1', '#FFC8DD', '#FFDFD3', '#FFF1AF']
)

plt.title("Deaths per 100,000 Population by Region", fontsize=14)
plt.tight_layout()
plt.show()
```

## Additional analysis

### 1) UNIVARIATE ANALYSIS

```python
deaths = df["Estimated number of deaths from TB (all forms, excluding HIV)"]
plt.figure(figsize=(9,5))
plt.hist(deaths, bins=40, density=True, alpha=0.4)
density = np.histogram(deaths, bins=40, density=True)
plt.plot(np.linspace(min(deaths), max(deaths), 40), density[0], linewidth=2)
plt.title("Smooth Distribution of TB Deaths", fontsize=14)
plt.xlabel("Deaths")
plt.ylabel("Density")
plt.tight_layout()
plt.show()
```

```python
region_counts = df["Region"].value_counts()

plt.figure(figsize=(5,5))
plt.pie(region_counts, labels=region_counts.index, autopct="%1.1f%%",
        pctdistance=0.85, startangle=140)
centre = plt.Circle((0,0),0.60,fc='white')
plt.gca().add_artist(centre)
plt.title("Regional Representation in Dataset", fontsize=14)
plt.tight_layout()
plt.show()
```

### 2) BIVARIATE ANALYSIS

```python
# Auto-detect the incidence per 100k column
inc_col = [col for col in df.columns 
           if "incidence" in col.lower() and "100" in col.lower() and "all forms" in col.lower()]

print("Detected incidence column:", inc_col[0])

# Now plot using the detected column
plt.figure(figsize=(10,6))

plt.scatter(
    df["Estimated total population number"],
    df["Estimated number of deaths from TB (all forms, excluding HIV)"],
    s=df[inc_col[0]] * 0.5,
    alpha=0.6
)

plt.title("Population vs Deaths (Bubble Size = Incidence Rate per 100k)", fontsize=14)
plt.xlabel("Population")
plt.ylabel("Deaths")
plt.tight_layout()
plt.show()
```

```python
regions = df["Region"].unique()

plt.figure(figsize=(12,7))

data = [df[df["Region"]==r]["Estimated number of deaths from TB (all forms, excluding HIV)"].dropna()
        for r in regions]

plt.boxplot(data, labels=regions, patch_artist=True)

plt.title("Distribution of TB Deaths by Region", fontsize=14)
plt.xlabel("Region")
plt.ylabel("TB Deaths")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
region_col = "Region"
deaths_col = "Estimated number of deaths from TB (all forms, excluding HIV)"

# Violin plot
plt.figure(figsize=(8,5))
sns.violinplot(
    x=region_col,
    y=deaths_col,
    data=df,
    palette="Set2",
    inner="quartile"
)

plt.title("Distribution of TB Deaths Across Regions (Violin Plot)", fontsize=14)
plt.xlabel("Region", fontsize=12)
plt.ylabel("TB Deaths", fontsize=12)
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

### 3) MULTIVARIATE ANALYSIS

```python

num_df = df.select_dtypes(include=[np.number])
corr = num_df.corr()

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

plt.figure(figsize=(18,12))

sns.heatmap(
    corr,
    mask=mask,
    cmap="viridis",
    annot=False,
    linewidths=0.5,
    cbar_kws={"shrink": 0.6}
)

plt.title("Correlation Heatmap of TB Metrics", fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.yticks(rotation=0, fontsize=8)

plt.tight_layout()
plt.show()

```

```python
top_regions = df["Region"].value_counts().nlargest(6).index
df_facet = df[df["Region"].isin(top_regions)]
g = sns.lmplot(
    x=inc_col,
    y="deaths_per_100k",
    col="Region",
    data=df_facet,
    col_wrap=3,
    height=4,
    scatter_kws={"alpha":0.6, "s":30},
    line_kws={"color":"black", "alpha":0.7}
)
g.set_axis_labels("Incidence per 100k", "Deaths per 100k")
g.fig.suptitle("Incidence vs Deaths per 100k (Faceted by Region)", fontsize=14, y=1.03)
plt.show()
```
